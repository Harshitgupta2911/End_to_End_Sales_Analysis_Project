```python
df['YearMonth'] = df['order_date'].dt.to_period('M').astype(str)
monthly_sales = (
    df.groupby('YearMonth')['sales']
      .sum()
      .reset_index()
)

plt.figure(figsize=(14,5))

sns.lineplot(
    data=monthly_sales,
    x='YearMonth',
    y='sales',
    marker='o'
)

plt.xticks(rotation=90)
plt.title("Monthly Sales Trend")
plt.xlabel("Year-Month")
plt.ylabel("Total Sales")
plt.tight_layout()

plt.show()
```

```python
df['month_num'] = df['order_date'].dt.month
```

```python
df['YearQuarter'] = df['order_date'].dt.to_period('Q').astype(str)
monthly_sales = (
    df.groupby('YearQuarter')['sales']
      .sum()
      .reset_index()
)

plt.figure(figsize=(14,5))

sns.lineplot(
    data=monthly_sales,
    x='YearQuarter',
    y='sales',
    marker='o'
)

plt.xticks(rotation=90)
plt.title("Quarterly Sales Trend")
plt.xlabel("Year-Quarter")
plt.ylabel("Total Sales")
plt.tight_layout()

plt.show()
```

```python
yearly_sales = (
    df.groupby('year')['sales']
      .sum()
      .reset_index()
)
plt.figure(figsize=(14,5))
sns.lineplot(
    data=yearly_sales,
    x='year',
    y='sales',
    marker='o'
)

plt.title("Yearly Sales Trend")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.tight_layout()
```

```python
from sqlalchemy import create_engine
from sqlalchemy.engine import URL

# PostgreSQL Credentials
username = "postgres"
password = "Maths@29115"      # Replace with your password
host = "localhost"
port = 5432
database = "Sales_Analytics"

# Create connection URL
url = URL.create(
    drivername="postgresql+psycopg2",
    username=username,
    password=password,
    host=host,
    port=port,
    database=database
)

# Create engine
engine = create_engine(url)
```

```python
try:
    with engine.connect() as conn:
        print("✅ Connected to PostgreSQL successfully!")
except Exception as e:
    print("❌ Connection failed!")
    print(e)
```

```python
df.to_sql(
    name='sales',
    con=engine,
    if_exists='replace',
    index=False
)

print("✅ Data uploaded successfully!")
```
